```python
## Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
## Load files
def load_and_prepare_data(folder_path):
    """Loads and prepares data from CSV files in a given folder."""
    data_frames = {}
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".csv"):
            instrument_name = file_name.split(".csv")[0]
            file_path = os.path.join(folder_path, file_name)
            df = pd.read_csv(file_path)
            data_frames[instrument_name] = df
    return data_frames
folder_path = 'RawDataDoNotModify'
data_frames = load_and_prepare_data(folder_path)
data_frames.keys()
```

```python
# Calculate moving averages and ATR
def summarizing_market_conditions(data):
    data['SMA_20'] = data['Close'].rolling(window=20).mean()
    
    # Calculate Average True Range (ATR) for volatility filtering
    data['High-Low'] = data['High'] - data['Low']
    data['High-Close'] = np.abs(data['High'] - data['Close'].shift(1))
    data['Low-Close'] = np.abs(data['Low'] - data['Close'].shift(1))
    data['TrueRange'] = data[['High-Low', 'High-Close', 'Low-Close']].max(axis=1)
    data['ATR'] = data['TrueRange'].rolling(window=14).mean()

    # Define conditions
    def categorize_market(row):
        if pd.isna(row['SMA_20']):  # Avoid NA in first rows
            return np.nan
        if abs(row['Close'] - row['SMA_20']) <= 0.01 * row['SMA_20']:  # Ranging
            return 'Ranging'
        elif row['ATR'] > 0.02 * row['Close']:  # Choppy
            return 'Choppy'
        else:  # Trending
            return 'Trending'

    data['Market_Condition'] = data.apply(categorize_market, axis=1)

    # Calculate time spent in each condition
    time_spent = data['Market_Condition'].value_counts(normalize=True).dropna()

    # Calculate lengths and levels for each market condition
    market_summary = data.groupby('Market_Condition').agg(
        Duration=('Market_Condition', 'size'),  # Count of occurrences
        Average_Level=('Close', 'mean'),  # Average closing price
    )

    # Add percentage of total duration for each condition
    market_summary['Percentage'] = (market_summary['Duration'] / len(data)) * 100

    # Reset index for a cleaner output
    market_summary = market_summary.reset_index()

    print(f"{key} -> Market Condition Summary:")
    print(market_summary)

for key, value in data_frames.items():
    summarizing_market_conditions(data_frames[key])
    print("------------------")
```

## Question-2

```python
list_of_assets = ['XAUUSD_Candlestick_1_Hour_BID_01.01.2020-31.08.2024',
                 'EURUSD_Candlestick_1_Hour_BID_01.01.2020-31.08.2024',
                 'USA500.IDXUSD_Candlestick_1_Hour_BID_01.01.2020-31.08.2024']
mini_dataframes = {}
for i in list_of_assets:
    mini_dataframes[i] = data_frames[i]
mini_dataframes.keys()
```

```python
def strategy_with_sharpe_and_highest_return(data):
    # Calculate daily returns
    data['Returns'] = data['Close'].pct_change()
    
    # Define Moving Averages
    short_window = 10
    long_window = 30
    data['MA10'] = data['Close'].rolling(window=short_window).mean()
    data['MA30'] = data['Close'].rolling(window=long_window).mean()

    # Calculate Average True Range (ATR) for volatility filtering
    data['High-Low'] = data['High'] - data['Low']
    data['High-Close'] = np.abs(data['High'] - data['Close'].shift(1))
    data['Low-Close'] = np.abs(data['Low'] - data['Close'].shift(1))
    data['TrueRange'] = data[['High-Low', 'High-Close', 'Low-Close']].max(axis=1)
    data['ATR'] = data['TrueRange'].rolling(window=14).mean()
    
    # Define a function for backtesting with exit strategy and ATR filter
    def backtest_ma_strategy(data, exit_bars, atr_threshold):
        data = data.copy()
        data['Position'] = 0
        data.loc[data.index[short_window:], 'Position'] = np.where(
            (data['MA10'][short_window:] > data['MA30'][short_window:]) & 
#             (data['ATR'][short_window:] > atr_threshold), 1, -1
            (data['ATR'][short_window:] > data['ATR'][short_window:]*atr_threshold), 1, -1
        )

        # Exit after `exit_bars` periods if no new signal
        for i in range(short_window, len(data)):
            if data['Position'].iloc[i] in [1, -1]:
                exit_index = min(i + exit_bars, len(data) - 1)
                data.loc[data.index[i+1:exit_index], 'Position'] = np.nan  # Clear future positions to apply exit

        data['Position'] = data['Position'].ffill().fillna(0)
        data['Strategy'] = data['Position'].shift(1) * data['Returns']

        # Calculate the number of trades (counting position changes)
        trades = data['Position'].diff().abs().sum() / 2  # Divide by 2 as each trade includes both entry and exit

        # Calculate Sharpe ratio
        strategy_returns = data['Strategy'].dropna()
        sharpe_ratio = strategy_returns.mean() / strategy_returns.std() * np.sqrt(252)  # Annualized Sharpe ratio

        # Calculate cumulative return
        cumulative_return = (1 + strategy_returns).prod() - 1  # Total cumulative return for the strategy

        return sharpe_ratio, int(trades), cumulative_return
    
    
    # Define a range of parameters
    exit_bars_values = [1, 3, 5, 10, 15, 20]
    atr_threshold_values = [0.0001, 0.0001, 0.001, 1]
    
    # Store results
    results = []

    for exit_bars in exit_bars_values:
        for atr_threshold in atr_threshold_values:
            sharpe_ratio, trades, cumulative_return = backtest_ma_strategy(data, exit_bars, atr_threshold)
            
            # Format and round values for readability
            sharpe_ratio = round(sharpe_ratio, 3) if sharpe_ratio > 0.01 else round(sharpe_ratio, 6)
            cumulative_return = round(cumulative_return, 4)
            
            results.append((exit_bars, atr_threshold, sharpe_ratio, trades, cumulative_return*100))

    # Convert to DataFrame for analysis
    results_df = pd.DataFrame(results, columns=['Exit Bars', 'ATR Threshold', 'Sharpe Ratio', 'Trades', 'Cumulative Return'])

    # Find the highest cumulative return and corresponding parameters
    highest_return_row = results_df.loc[results_df['Cumulative Return'].idxmax()]

    # Print the highest return and associated parameters
    print("Highest Return:")
    print(highest_return_row)

    # Plot heatmap for Sharpe Ratio
    
    pivot = results_df.pivot("Exit Bars", "ATR Threshold", "Sharpe Ratio")
    plt.figure(figsize=(10, 8))
    sns.heatmap(pivot, annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Sharpe Ratio Heatmap")
    plt.xlabel("ATR Threshold")
    plt.ylabel("Exit Bars")
    plt.show()

# Example usage:
# Replace `data` with your actual DataFrame containing 'Close', 'High', and 'Low'
# strategy_with_sharpe_and_highest_return(data)
```

```python
for key, value in mini_dataframes.items():
    name = key.split("_")
    print(name[0])
    strategy_with_sharpe_and_highest_return(mini_dataframes[key])
```

## Strategy Improvement

```python
from scipy.stats import linregress

def strategy_with_regression_filter(data):
    # Calculate daily returns
    data['Returns'] = data['Close'].pct_change()
    
    # Define Moving Averages
    short_window = 10
    long_window = 30
    data['MA10'] = data['Close'].rolling(window=short_window).mean()
    data['MA30'] = data['Close'].rolling(window=long_window).mean()

    # Calculate Average True Range (ATR) for volatility filtering
    data['High-Low'] = data['High'] - data['Low']
    data['High-Close'] = np.abs(data['High'] - data['Close'].shift(1))
    data['Low-Close'] = np.abs(data['Low'] - data['Close'].shift(1))
    data['TrueRange'] = data[['High-Low', 'High-Close', 'Low-Close']].max(axis=1)
    data['ATR'] = data['TrueRange'].rolling(window=14).mean()

    # Calculate Linear Regression Slope
    regression_window = 20  # Number of periods for regression
    data['Slope'] = data['Close'].rolling(window=regression_window).apply(
        lambda x: linregress(range(len(x)), x).slope, raw=False
    )

    # Define a function for backtesting with regression filter
    def backtest_ma_strategy_with_filter(data, exit_bars, atr_threshold, slope_threshold):
        data = data.copy()
        data['Position'] = 0
        data.loc[data.index[long_window:], 'Position'] = np.where(
            (data['MA10'][long_window:] > data['MA30'][long_window:]) & 
            (data['ATR'][long_window:] > data['ATR'][long_window:]*atr_threshold) & 
            (data['Slope'][long_window:].abs() > slope_threshold), 1, -1
        )

        # Exit after `exit_bars` periods if no new signal
        for i in range(long_window, len(data)):
            if data['Position'].iloc[i] in [1, -1]:
                exit_index = min(i + exit_bars, len(data) - 1)
                data.loc[data.index[i+1:exit_index], 'Position'] = np.nan

        data['Position'] = data['Position'].ffill().fillna(0)
        data['Strategy'] = data['Position'].shift(1) * data['Returns']

        # Calculate performance metrics
        trades = data['Position'].diff().abs().sum() / 2
        strategy_returns = data['Strategy'].dropna()
        sharpe_ratio = strategy_returns.mean() / strategy_returns.std() * np.sqrt(252)
        cumulative_return = (1 + strategy_returns).prod() - 1

        return sharpe_ratio, int(trades), cumulative_return
    
    # Define parameter ranges
    exit_bars_values = [3, 5, 10, 15, 20]
    atr_threshold_values = [0.0001, 0.0001, 0.001, 1]
    slope_threshold_values = [0.01, 0.02, 0.03]  # Adjust based on your data's scale

    # Store results
    results = []
    for exit_bars in exit_bars_values:
        for atr_threshold in atr_threshold_values:
            for slope_threshold in slope_threshold_values:
                sharpe_ratio, trades, cumulative_return = backtest_ma_strategy_with_filter(
                    data, exit_bars, atr_threshold, slope_threshold
                )
                results.append((exit_bars, atr_threshold, slope_threshold, sharpe_ratio, trades, cumulative_return*100))

    # Convert results to DataFrame
    results_df = pd.DataFrame(results, columns=[
        'Exit Bars', 'ATR Threshold', 'Slope Threshold', 
        'Sharpe Ratio', 'Trades', 'Cumulative Return'
    ])

    # Find the highest cumulative return and corresponding parameters
    highest_return_row = results_df.loc[results_df['Cumulative Return'].idxmax()]

    # Print the highest return and associated parameters
    print("Highest Return:")
    print(highest_return_row)

    # Plot Sharpe Ratio Heatmap for Exit Bars and ATR Threshold
    results_df = results_df.drop_duplicates(subset=['Exit Bars', 'ATR Threshold'])
    pivot = results_df.pivot("Exit Bars", "ATR Threshold", "Sharpe Ratio")
    plt.figure(figsize=(10, 8))
    sns.heatmap(pivot, annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Sharpe Ratio Heatmap with Regression Filter")
    plt.xlabel("ATR Threshold")
    plt.ylabel("Exit Bars")
    plt.show()

# Example usage:
# Replace `data` with your actual DataFrame containing 'Close', 'High', 'Low'
# strategy_with_regression_filter(data)
```

### **Report: Enhancing Strategy with Regression Slope Filter**

---

#### **1. Performance Metrics for MA Cross Over + ATR Each Market**
- MA + ATR strategy worked well on Gold because, MA crossover strategy is trendbased strategy. XADUSD is a unidirectional trending market(either up or down). 

#### **3. Explanation of the Chosen Filter**
- **Filter Used**: **Regression Slope**
  - The slope of a linear regression line was calculated over a rolling window of 20 periods.
  - It acts as a trend strength indicator:  
    - **Positive Slope**: Uptrend.  
    - **Negative Slope**: Downtrend.
  - This filter gives more power to the trend enhancing treding markets more. The filter eliminated trades in choppy or weakly trending markets, reducing unnecessary losses.

#### **4. Insights and Conclusions**
- As expected it increased returns on XADUSD(Gold). Improved by 4% in markets with strong trends. And performance is -ve on other two markets. 

**4.3. Key Insights**
- The strategy performs best in strongly trending markets. The regression slope filter enhanced the ability to capture such trends.
- The filter reduced the number of trades, improving efficiency and reducing transaction costs.

The strategy is best suited for markets like **XADUSD**, where the filter capitalized on strong and consistent trends. Future work could involve testing additional trend-based filters or optimizing parameters for specific market conditions.
