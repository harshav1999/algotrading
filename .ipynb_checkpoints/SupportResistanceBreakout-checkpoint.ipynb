```python
import pandas as pd
```

```python
df = pd.read_csv('RawDataDoNotModify/BTCUSD_Candlestick_1_Hour_BID_01.01.2020-31.08.2024.csv')
## Remove rows with no volume
df = df[df['Volume']!=0]
df = df.tail(10000)
df.reset_index(drop=True, inplace=True)
```

```python
df.head(4)
```

```python
## Identify Support and Resistance

## 1. Indetify pivot points
def pivotid(df1, l, n1, n2): #n1 n2 before and after candle l
    if l-n1 < 0 or l+n2 >= len(df1):
        return 0
    
    pividlow=1
    pividhigh=1
    
    for i in range(l-n1, l+n2+1):
        if(df1.Low[l]>df1.Low[i]):
            pividlow=0
        if(df1.High[l]<df1.High[i]):
            pividhigh=0
    if pividlow and pividhigh:
        return 3
    elif pividlow:
        return 1
    elif pividhigh:
        return 2
    else:
        return 0
    
df['pivot'] = df.apply(lambda x: pivotid(df, x.name,10,10), axis=1)
```

```python
import numpy as np
def pointpos(x):
    if x['pivot']==1:
        return x['Low']-1e-3
    elif x['pivot']==2:
        return x['High']+1e-3
    else:
        return np.nan

df['pointpos'] = df.apply(lambda row: pointpos(row), axis=1)
```

```python
# Define the function to calculate supports and resistances
def calculate_support_resistance(df, price_range, x_candles):
    # Add new columns for support and resistance based on the 'pivot' column
    df['support'] = df.apply(lambda row: row['pointpos'] if row['pivot'] == 1 else None, axis=1)
    df['resistance'] = df.apply(lambda row: row['pointpos'] if row['pivot'] == 2 else None, axis=1)
    
    # Aggregate support and resistance within the specified price range
    supports = df['support'].dropna()
    resistances = df['resistance'].dropna()
    
    def aggregate_points(points, price_range):
        averaged_points = []
        while len(points) > 0:
            current = points.iloc[0]
            close_points = points[abs(points - current) <= price_range]
            averaged_points.append(close_points.mean())
            points = points[abs(points - current) > price_range]
        return averaged_points
    
    avg_supports = aggregate_points(supports, price_range)
    avg_resistances = aggregate_points(resistances, price_range)
    
    # Find the closest aggregated support and resistance for each candle
    closest_supports = []
    closest_resistances = []
    
    for i in range(len(df)):
        past_data = df.iloc[max(0, i-x_candles):i]  # Get past x_candles
        if not past_data.empty:
            closest_supports.append(
                min(avg_supports, key=lambda s: abs(s - df.loc[i, 'Open'])) if avg_supports else None
            )
            closest_resistances.append(
                min(avg_resistances, key=lambda r: abs(r - df.loc[i, 'High'])) if avg_resistances else None
            )
        else:
            closest_supports.append(None)
            closest_resistances.append(None)
    
    df['closest_support'] = closest_supports
    df['closest_resistance'] = closest_resistances
    
    return df

# Apply the function with a price range of 500 and past 5 candles
result_df = calculate_support_resistance(df, price_range=500, x_candles=5)
```
